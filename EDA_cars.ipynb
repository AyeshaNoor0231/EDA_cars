```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes = True)
```

```python
df = pd.read_csv("/content/data_Lec13.csv")
df.head(10)
```

```python
df.tail(5)
```

```python
df.dtypes
```

# Dropping values

```python
df = df.drop(["Vehicle Size"], axis = 1) #axis = 0 drop values along rows, axis=1 drop along columns
df.head(5)
```

# Renaming colums

```python
df = df.rename(columns = {"Engine HP" : "HP", "Engine Cylinders":"Cylinders"})
df.head(5)
```

```python
df.shape
```

```python
duplicate = df[df.duplicated()]
duplicate.shape
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.head(4)
```

```python
df.count()
```

**Dropping the missing values or Null Values**

```python
print(df.isnull().sum())
```

```python
df= df.dropna()
df.count()
```

```python
print(df.isnull().sum())
```

# Detecting Outliers

```python
sns.boxplot(x=df['HP'])
```

```python
sns.boxplot(x=df['Cylinders'])
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
sns.boxplot(y= df['Vehicle Style'],x=df['Cylinders']) #seaborn boxplot work for numerical values
```

```python
sns.countplot(x= df['Vehicle Style'])
plt.xticks(rotation = 90) # for better visualization
plt.show()
```

**Correlation to find which parameters are more importanr, by checking their correlation. +1 to -1. The more the value is near to 1, the stronger the relation in either positive or negative direction. Sign shows direction of influence**

```python
df.corr()
```

```python
df.Make.value_counts().nlargest(20).plot(kind='bar', figsize=(10,5))
plt.title("Price of cars by make")
plt.ylabel('price of cars')
plt.xlabel('Make');
```

```python
df.Make.value_counts().nlargest(10).plot(kind = 'area', figsize = (10,5))
plt.ylabel('Price')
plt.xlabel('Make')
plt.show()
```

```python
df.Year.value_counts().plot(kind = 'pie', figsize=(15,10))
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize= (10,5))
sns.heatmap(df.corr(), cmap = 'coolwarm', annot = True, vmin = -1, vmax= +1)
```
